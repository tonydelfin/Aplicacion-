## Ejercicio

Escribir un programa que pregunte el nombre del usuario en la consola y un número entero e imprima por pantalla en líneas distintas el nombre del usuario tantas veces como el número introducido.

## Solución

```python

import math
INGREDIENTES = [
    'Manzana',
    'Avena',
    'Chía',
    'Semillas de girasol',
    'Miel',
    'Canela',
    'Vainilla',
    'Royal'
]

CANTIDADES_BASE = [3, 1, 2, 2, 3, 1, 1, 1]

PRECIOS = [4.0, 8.0, 12.0, 10.0, 15.0, 3.0, 5.0, 2.0]

PORCIONES_BASE = 12

def escalar_receta(porciones):
    """
    Escala las cantidades de ingredientes según las porciones deseadas.
    """
    factor = porciones / PORCIONES_BASE
    return [cantidad * factor for cantidad in CANTIDADES_BASE]

def calcular_costo(cantidades, margen=30):
    """
    Calcula el costo base y el precio sugerido con margen de ganancia.
    """
    costo = sum(cantidades[i] * PRECIOS[i] for i in range(len(INGREDIENTES)))
    precio = costo * (1 + margen / 100)
    return costo, precio

print("RECETA BASE DE GALLETAS DE VAINILLA")
for i in range(len(INGREDIENTES)):
    print(f"{INGREDIENTES[i]}: {CANTIDADES_BASE[i]}")

cantidades = escalar_receta(48)
costo, precio = calcular_costo(cantidades)

print(f"\nCosto para 48 galletas: ${costo:.2f}")
print(f"Precio sugerido con margen: ${precio:.2f}")
```

```python
TIEMPO_HORNO = [0, 3, 6, 9, 12, 15, 18]
TEMP_INTERIOR = [24, 75, 125, 155, 175, 182, 185]

def integracion_trapecio(x, y):
    """
    Calcula el área bajo la curva usando el método del trapecio.
    """
    area = 0

    for i in range(len(x) - 1):
        h = x[i + 1] - x[i]
        area += h * (y[i] + y[i + 1]) / 2

    return area

Q_trap = integracion_trapecio(TIEMPO_HORNO, TEMP_INTERIOR)
masa = 0.035
calor_especifico = 2.3
calor = masa * calor_especifico * Q_trap / 60

print("\nINTEGRACIÓN POR TRAPECIO")
print(f"Área: {Q_trap:.2f} °C·min")
print(f"Calor aproximado por galleta: {calor:.2f} kJ")
```

```python
def integracion_simpson13(x, y):
    """
    Calcula el área bajo la curva usando Simpson 1/3.
    Si el número de intervalos es impar, usa trapecio en el último tramo.
    """
    n = len(x) - 1
    h = (x[-1] - x[0]) / n

    if n % 2 == 0:
        area = y[0] + y[-1]

        for i in range(1, n, 2):
            area += 4 * y[i]

        for i in range(2, n, 2):
            area += 2 * y[i]

        return area * h / 3

    else:
        area_simpson = integracion_simpson13(x[:-1], y[:-1])
        area_trapecio = (x[-1] - x[-2]) * (y[-2] + y[-1]) / 2
        return area_simpson + area_trapecio

Q_simp = integracion_simpson13(TIEMPO_HORNO, TEMP_INTERIOR)

print("\nINTEGRACIÓN POR SIMPSON")
print(f"Trapecio: {Q_trap:.2f} °C·min")
print(f"Simpson: {Q_simp:.2f} °C·min")
print(f"Diferencia: {abs(Q_trap - Q_simp):.2f} °C·min")
```

```python
k = 0.11
H0 = 28.0
tf = 18
h_euler = 1.0

def dH_dt(t, H):
    """
    Modelo de pérdida de humedad en galletas.
    H representa el porcentaje de humedad.
    """
    return -k * H

def euler(f, y0, t0, tf, h):
    """
    Resuelve una EDO usando el método de Euler.
    """
    t = [t0]
    y = [y0]

    while t[-1] < tf:
        y.append(y[-1] + h * f(t[-1], y[-1]))
        t.append(t[-1] + h)

    return t, y

t_e, H_e = euler(dH_dt, H0, 0, tf, h_euler)
exact = H0 * math.exp(-k * tf)

print("\nMÉTODO DE EULER — PÉRDIDA DE HUMEDAD")
print(f"Euler H(18) = {H_e[-1]:.2f}%")
print(f"Exacto aproximado = {exact:.2f}%")
print(f"Error = {abs(H_e[-1] - exact):.2f}%")
```

```python
def rk4(f, y0, t0, tf, h):
    """
    Resuelve una EDO usando Runge-Kutta de cuarto orden.
    """
    t = [t0]
    y = [y0]

    while t[-1] < tf:
        ti = t[-1]
        yi = y[-1]

        k1 = f(ti, yi)
        k2 = f(ti + h / 2, yi + h * k1 / 2)
        k3 = f(ti + h / 2, yi + h * k2 / 2)
        k4 = f(ti + h, yi + h * k3)

        y.append(yi + h * (k1 + 2*k2 + 2*k3 + k4) / 6)
        t.append(ti + h)

    return t, y

t_rk, H_rk = rk4(dH_dt, H0, 0, tf, h_euler)

print("\nRK4 VS EULER — HUMEDAD EN GALLETAS")
print(f"RK4 H(18) = {H_rk[-1]:.4f}%")
print(f"Euler H(18) = {H_e[-1]:.2f}%")
print(f"Exacto aproximado = {exact:.4f}%")
print(f"Error RK4 = {abs(H_rk[-1] - exact):.8f}")
print(f"Error Euler = {abs(H_e[-1] - exact):.4f}")
```

```python
def sistema_galletas(t, Y):
    """
    Sistema acoplado:
    T = temperatura interna de la galleta
    H = humedad de la galleta

    La temperatura aumenta por el horno y la humedad disminuye
    según la temperatura interna.
    """
    T, H = Y

    T_horno = 185.0
    alpha = 0.20
    beta = 0.06

    dT = alpha * (T_horno - T)
    dH = -beta * (T / 100) * H

    return [dT, dH]

def rk4_sistema(f, Y0, t0, tf, h):
    """
    Resuelve un sistema de EDOs usando RK4.
    """
    t = [t0]
    Y = [Y0[:]]

    while t[-1] < tf:
        ti = t[-1]
        Yi = Y[-1]

        k1 = f(ti, Yi)
        k2 = f(ti + h / 2, [Yi[j] + h * k1[j] / 2 for j in range(len(Yi))])
        k3 = f(ti + h / 2, [Yi[j] + h * k2[j] / 2 for j in range(len(Yi))])
        k4 = f(ti + h, [Yi[j] + h * k3[j] for j in range(len(Yi))])

        Ynuevo = [
            Yi[j] + h * (k1[j] + 2*k2[j] + 2*k3[j] + k4[j]) / 6
            for j in range(len(Yi))
        ]

        Y.append(Ynuevo)
        t.append(ti + h)

    return t, Y

Y0 = [24.0, 28.0]
t_sys, Y_sys = rk4_sistema(sistema_galletas, Y0, 0, 18, 1)

print("\nSISTEMA ACOPLADO — GALLETAS")
print(f"Temperatura final T(18) = {Y_sys[-1][0]:.2f} °C")
print(f"Humedad final H(18) = {Y_sys[-1][1]:.2f}%")
```

```python
R = 8.314
Ea = 42000
k_ref = 0.004
T_ref_k = 298.15

def vida_util_galletas(T_c):
    """
    Estima la vida útil de las galletas según la temperatura de almacenamiento.
    """
    T_k = T_c + 273.15
    k_T = k_ref * math.exp(-Ea / R * (1 / T_k - 1 / T_ref_k))
    dias = math.log(100 / 50) / (k_T * 24)
    return dias, k_T

print("\nPREDICTOR DE VIDA ÚTIL — GALLETAS")
for Tc in [15, 20, 25, 30, 35, 40]:
    dias, kT = vida_util_galletas(Tc)
    print(f"{Tc} °C → {dias:.1f} días")
```

```python
Ea_L = 70000
k_L_ref = 0.006
L0, L_inf = 82.0, 35.0
T_ref_c = 180

def color_galleta(T_c, t):
    """
    Estima el color de la galleta durante el horneado.
    L* alto = claro
    L* bajo = dorado/oscuro
    """
    k_L = k_L_ref * math.exp(-Ea_L / R * (1 / (T_c + 273.15) - 1 / (T_ref_c + 273.15)))
    return L_inf + (L0 - L_inf) * math.exp(-k_L * t)

print("\nCOLOR DE GALLETA")
print(f"{'t(min)':>8} {'T(°C)':>10} {'L*':>10}")

for i in [0, 3, 6, 9, 12, 15, 18]:
    T_i = Y_sys[min(i, len(Y_sys) - 1)][0]
    L = color_galleta(T_i, i)
    print(f"{i:>8} {T_i:>10.2f} {L:>10.2f}")
```
